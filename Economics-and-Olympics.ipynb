### GDP vs Total Medals
First we will examine the relationship between GDP and the amount of medals earned for each country that has an available 2019 GDP (PPP). This means that groups that have not yet earned a medal, are not considered countries, or are countries for which their 2019 GDP was not easily available were not included in this scatter plot.

A good explanation of GDP (PPP) provided by Wikipedia is that
    
   “GDP comparisons using PPP are arguably more useful than those using nominal GDP when assessing a nation's domestic market because PPP takes into account the relative cost of local goods, services and inflation rates of the country, rather than using international market exchange rates, which may distort the real differences in per capita income.”

```python
# reads the csv files and assigns their contents to medals
# and countries respectively
medals = pd.read_csv('all_medals.csv');
GDPs = pd.read_csv('GDP_PPP.csv');

# get the x and y values for each of the plots
[GDP1, medal1] = plot_MedalsvGDP(medals, GDPs);
[GDP2, medal2] = plot_MedalsvGDP_noOutliers(medals, GDPs);

# perform statistical hypothesis test on the variables
find_Spearman_coefficient(GDP1, medal1)
find_Spearman_coefficient(GDP2, medal2)
```

The first (top) graph contains 3 clear outliers that are the U.S, Russia and India, the countries with the highest GDPs in 2019. When they are removed in the second graph, we see clearly that there is a cluster of countries with low GDPs and low medal counts. With the spearman coefficient being 0.672 and 0.660 with and without outliers respectively, we start to see a connection starting to form between economic well-being and medal 
count.

### GDP (PPP) per Capita vs Total Medals

The third pair of scatter plots in this essay is the GDP (PPP) per Capita (takes the GDP of a country and divides it by its population) vs. Total Medals.

While  finding the GDP per Capita form the GDP is a simple math operation, the difference between these two measures in real life is neither simple, nor is it trivial. While GDP is a popular method to measure economic status, population size is not something that can be ignored. By dividing the GDP by the population, we can get a better feel for how well-off every citizen would be if the wealth of a nation was distributed equally, and gives a clearer picture of the quality of life and living standards of the citizens.

If only GDP was used to measure economic activity, a country with a high GDP but very high population would be considered economically better off than a small country with a smaller GDP, which we know in real life is not always the case. GDP is influenced, and therefore can be skewed by very large or very small populations, whilc GDP per Capita takes into account this factor to provide a more accurate sense of a country's eceonmic well-being.


```python
# reads the csv files and assigns their contents to medals
# and GDPs per capita respectively
medals = pd.read_csv('all_medals.csv');
GDPs = pd.read_csv('GDP_PPP_per_Capita.csv');

# get the x and y values for each of the plots
[GDP1, medal1] = plot_MedalsvGDP_per_Capita(medals, GDPs);
[GDP2, medal2] = plot_MedalsvGDP_per_Capita_noOutliers(medals, GDPs);

# perform statistical hypothesis test on the variables
find_Spearman_coefficient(GDP1, medal1)
find_Spearman_coefficient(GDP2, medal2)
```

Again, the outlier is the USA with significantly more medals than anyone else.It’s correlation coefficient here is 0.469. Without the USA, we see the correlation drop to 0.458; both correlations are still in the moderate range.


### IHDI vs Total Medals

Inequality-adjusted Human Development Index   defined as the level of human development in a country based on the life expectancy, education, and per capita income levels of its citizens, with inequality taken into account, unlike the HDI. We decided to also include the IHDI, as it is a bit different from the GDP and GDP per Capita. Those two measures focus heavily on output and the monetary value of that output, but IHDI considers not only per capita income and money, but also life expectancy and education.

```python
# reads the csv files and
#  assigns their contents to medals
# and IHDIs per capita respectively
medals = pd.read_csv('all_medals.csv');
IHDIs = pd.read_csv('IHDIs.csv');

# get the x and y values for each of the plots
[GDP1, medal1] = plot_MedalsvIHDI(medals, IHDIs);
[GDP2, medal2] = plot_MedalsvIHDI_noOutliers(medals, IHDIs);

# perform statistical hypothesis 
# test on the variables
find_Spearman_coefficient(GDP1, medal1)
find_Spearman_coefficient(GDP2, medal2)
```

In the first graph shown, the United States is the sole outlier with its unusually high medal count that throws off the shape of the graph. The second plot with the outlier gone has a clearer shape in that we can see and the beginnings of an exponential growth curve. The Spearman's coefficient for the graph with outliers, is 0.552, and the coefficient without outliers is 0.548.With these two coefficient calues, which indicate moderate correlation we have further evidence of the relationship between the two variables.
